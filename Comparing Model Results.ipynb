```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
hybrid_df = pd.read_csv('Result of Hybrid Strategy.csv')
over_df = pd.read_csv('Result of Oversampling Strategy.csv')
under_df = pd.read_csv('Result of Undersampling Strategy.csv')
```

```python
df_models = pd.DataFrame(under_df['Models'])
df_models
```

### Comparing the model results of the Hybrid and the Oversampling Strategy

```python
df_hvo =  hybrid_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']] > over_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']]
df_hvo = pd.concat([df_models, df_hvo], axis = 1)
df_hvo
```

### Comparing the model results of the Hybrid and the Undersampling Strategy

```python
df_huo = hybrid_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']] > under_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']]
df_huo = pd.concat([df_models, df_huo], axis = 1)
df_huo
```

### Comparing the model results of the Oversampling and the Undersampling Strategy

```python
df_ou = over_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']] > under_df[['Accuracy','Precision','Recall','f1','ROC_AUC','Average Precision']]
df_ou = pd.concat([df_models, df_ou], axis = 1)
df_ou
```

### Comparing Model Training Time 

```python
df_time = pd.DataFrame({'Hybrid': ["2min 12s", '6min 46s', '4.89 s', '1min 28s', '2min 16s'],
                        'Oversampling': ["4min 55s", "1min 21s", '8.8 s', '2min 4s', '4min 38s'],
                        'Undersampling': ["375 ms", "3.4 s", '2.54 s', '21.3 s', '1min 5s']})
df_time = pd.concat([df_models, df_time], axis = 1)
df_time
```
